```python
from transformers import pipeline
```

```python
generator=pipeline("text-generation", model="distilgpt2")
generator.tokenizer.pad_token_id=generator.model.config.eos_token_id
generator.model.config.max_length=None
```

```python
print("AI Chatbot (type 'exit' to stop)\n")
while(True):
  user_input=input("YOU: ")
  if user_input.lower()=='exit':
    print("Bot: GoodBye!")
    break

  prompt="Your are a helpful assitant.\nUser:" + user_input + "\nBot: "

  response=generator(
        prompt,
        max_new_tokens=50,
        temperature=0.7,
        do_sample=True,
        top_p=0.9,
        repetition_penalty=1.2,


    )

  reply= response[0]['generated_text']#.split("Bot: ")[-1].strip()
  print("Bot: ",reply)
```
